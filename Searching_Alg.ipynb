Algoritma BFS ini akan menggunakan antrian (queue) 

Check the starting node and add its neighbours to the queue.

1. Tandai node mulai (starting node) - misal pada graph diatas A
2. Ambil node pertama dari antrian
3. Cek jika node sudah pernah di lewati 
4. Jika belum langsung menuju ke node tetangganya
5. Tambahkan tetangganya kedalam antrian
6. Tandai jika node telah di eksplore
7. Ulangi langkah 3 - 7 sampai antrian habis

Gambar diatas menggambarkan proses penempatan semua node ke dalam antrian dan pengambilan node dalam antrian yang direpresentasikan oleh graph sebelumnya. 
Representasi queue ini merupakan representasi graph untuk di antrikan di explore satu persatu 

Fungsi berikut digunakan untuk menunjukan urutan menyambangio graph dengan masukan node mulai (start) 

````
def bfs_connected_component(graph, start):
    # keep track of all visited nodes
    explored = []
    # keep track of nodes to be checked
    queue = [start]
 
    # keep looping until there are nodes still to be checked
    while queue:
        # pop shallowest node (first node) from queue
        node = queue.pop(0)
        if node not in explored:
            # add node to list of checked nodes
            explored.append(node)
            neighbours = graph[node]
 
            # add neighbours of node to queue
            for neighbour in neighbours:
                queue.append(neighbour)
    return explored
 
bfs_connected_component(graph,'A') # returns ['A', 'B', 'C', 'E', 'D', 'F', 'G']
````

Output yang di harapkan 

['A', 'B', 'C', 'E', 'D', 'F', 'G']

Fungsi di bawah ini digunakan untuk mencari jalur terpendek jika di berikan graph, node start dan node tujuannya  


````
# finds shortest path between 2 nodes of a graph using BFS
def bfs_shortest_path(graph, start, goal):
    # keep track of explored nodes
    explored = []
    # keep track of all the paths to be checked
    queue = [[start]]
 
    # return path if start is goal
    if start == goal:
        return "That was easy! Start = goal"
 
    # keeps looping until all possible paths have been checked
    while queue:
        # pop the first path from the queue
        path = queue.pop(0)
        # get the last node from the path
        node = path[-1]
        if node not in explored:
            neighbours = graph[node]
            # go through all neighbour nodes, construct a new path and
            # push it into the queue
            for neighbour in neighbours:
                new_path = list(path)
                new_path.append(neighbour)
                queue.append(new_path)
                # return path if neighbour is goal
                if neighbour == goal:
                    return new_path
 
            # mark node as explored
            explored.append(node)
 
    # in case there's no path between the 2 nodes
    return "So sorry, but a connecting path doesn't exist :("
 
bfs_shortest_path(graph, 'G', 'D')  # returns ['G', 'C', 'A', 'B', 'D']
````

output yang diharapkan 

['G', 'C', 'A', 'B', 'D']
