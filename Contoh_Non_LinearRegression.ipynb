Let's learn about non linear regressions and apply an example on python. In this notebook, we fit a non-linear model to the datapoints corrensponding to China's GDP from 1960 to 2014. *italicized text*)

<h2 id="importing_libraries">Mengimpor libraries</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Regresi linear bisa sangat baik untuk menyelsaikan masalah, meskipun tidak semua dataset cocok dengan regresi linear.
Contohnya ada persamaan regresi yang menunjukan hubungan antara x (bebas) dan y(terikat) dengan persamaan : y = $2x$ + 3.

Though Linear regression is very good to solve many problems, it cannot be used for all datasets. First recall how linear regression, could model a dataset. It models a linear relation between a dependent variable y and independent variable x. It had a simple equation, of degree 1, for example y = $2x$ + 3.

```python
sns.set(font_scale=1.3)
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph
y = 2*(x) + 3
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.figure(figsize=(8,6))
plt.suptitle('Grafik Regresi LInear')
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r')
plt.ylabel(' Variable Terikat')
plt.xlabel(' Variable Bebas')
plt.show()
```

Regresi Non-linear adalah hubungan antara variabel bebas  $x$ dan variabel terikat  $y$ yang tidak liner. Semua hubungan antara variabel terikat dan variabel bebas yang tidak linear disebut hubungan non linear. Dan biasanya dapat direpressentasikan dengan  polynomial $k$ derajat (maximum pangkat dari  $x$).

$$ \ y = a x^3 + b x^2 + c x + d \ $$

Fungsi Non-linear dapat meiliki elemen exponentials, logarithmik, pembagian, and lainya. Contoh : $$ y = \log(x)$$
    
atau yang lebih kompleks  :
$$ y = \log(a x^3 + b x^2 + c x + d)$$

### Cubic

Let's take a look at a cubic function's graph.

```python
x = np.arange(-5.0, 5.0, 0.1)
plt.figure(figsize=(8,6))
##You can adjust the slope and intercept to verify the changes in the graph
y = 1*(x**3) + 1*(x**2) + 1*x + 3
y_noise = 20 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r')
plt.suptitle('Cubic Regression Chart')
plt.ylabel(' Variable Terikat')
plt.xlabel('Variable Bebas')
plt.show()
```

Seperti yang dapat kita ligat fungsi memiliki  $x^3$ dan $x^2$ sebagai  variables bebas. ini adalah fungsi non-linear.

Fungsi non-linear antara lain :

### Kuadratik

# $$ Y-hat = X^2 $$

```python
x = np.arange(-5.0, 5.0, 0.1)
plt.figure(figsize=(8,6))
##You can adjust the slope and intercept to verify the changes in the graph

y = np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r')
plt.suptitle('Quadratic Regression Chart')
plt.ylabel('Variable Terikat')
plt.xlabel('Variable Bebas')
plt.show()
```

# \begin{equation}
y-hat = \log(X)

### Exponential

An exponential function with base c is defined by $$ Y = a + b c^X$$ where b 0, c > 0 , c 1, and x is any real number. The base, c, is constant and the exponent, x, is a variable.



```python
X = np.arange(-5.0, 5.0, 0.1)
plt.figure(figsize=(8,6))

##You can adjust the slope and intercept to verify the changes in the graph

Y= np.exp(X)
plt.plot(X,Y)
plt.suptitle('Exponential Regression Chart')
plt.ylabel(' Variable Terikat')
plt.xlabel('Variable Bebas')
plt.show()
```

### Logarithmic

The response $y$ is a results of applying logarithmic map from input $x$'s to output variable $y$. It is one of the simplest form of __log()__: i.e. $$ y = \log(x)$$

Please consider that instead of $x$, we can use $X$, which can be polynomial representation of the $x$'s. In general form it would be written as  
\begin{equation}
y = \log(X)
\end{equation}

```python
X = np.arange(-5.0, 5.0, 0.1)
plt.figure(figsize=(8,6))
Y = np.log(X)
plt.plot(X,Y)
plt.suptitle('Logarithmic Regression Chart')
plt.ylabel('Variable Terikat')
plt.xlabel('Variable Bebas')
plt.show()
```

### Sigmoidal/Logistic

# $$ Y = a + \frac{b}{1+ c^{(X-d)}}$$

```python
X = np.arange(-5.0, 5.0, 0.1)
plt.figure(figsize=(8,6))

Y = 1-4/(1+np.power(3, X-2))

plt.plot(X,Y)
plt.suptitle('Logistic Regression Chart')
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

# COntoh Regresi Non-Linear

Sebagai contoh kita akan mencoba mencari fungsi non linear dar data yang berasal dari data GDP china anyara tahun 1960 sampai 2014.

For an example, we're going to try and fit a non-linear model to the datapoints corresponding to China's GDP from 1960 to 2014. We download a dataset with two columns, the first, a year between 1960 and 2014, the second, China's corresponding annual gross domestic income in US dollars for that year.

```python
import numpy as np
import pandas as pd

#downloading dataset

china_gdp = 'https://raw.githubusercontent.com/Blackman9t/Machine_Learning/master/china_gdp.csv'

df = pd.read_csv(china_gdp)
df.head(10)
```

### Plotting the Dataset ###
This is what the datapoints look like. It kind of looks like an either logistic or exponential function. The growth starts off slow, then from 2005 on forward, the growth is very significant. And finally, it decelerate slightly in the 2010s.

```python
plt.figure(figsize=(8,5))
X_data, y_data = (df["Year"].values, df["Value"].values)
plt.plot(X_data, y_data, 'ro')
plt.suptitle('Graph menunjukan hubungan antara tahun dan GDP di CHINA', y=1.02)
plt.ylabel('GDP')
plt.xlabel('Tahun')
plt.show()
```

### Memilih  model ###

Dari penampakan grafik awal kita dapat menebak fungsi non linear yang cocok dengan trend pada data. Lalu misalnya kita memilih fungsi logistik yang jika di plot seperti gambar di bawah ini

From an initial look at the plot, we determine that the logistic function could be a good approximation,
since it has the property of starting with a slow growth, increasing growth in the middle, and then decreasing again at the end; as illustrated below: *italicized text*

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = 1.0 / (1.0 + np.exp(-X))

plt.plot(X,Y)
plt.ylabel(' Variable Terikat')
plt.xlabel('Variable Bebas')
plt.show()
```



Runus untuk   function logistic adalah:

$$ \hat{Y} = \frac1{1+e^{\beta_1(X-\beta_2)}}$$

$\beta_1$: Controls the curve's steepness,

$\beta_2$: Slides the curve on the x-axis.

### MEMBUAT  Model ###
Mari kita bangun model regresi dengan menginisialsiasi parameter beta1 dan beta 2

*Now, let's build our regression model and initialize its parameters.*

```python
def sigmoid(X, Beta_1, Beta_2):
     y = 1 / (1 + np.exp(-Beta_1*(X-Beta_2)))
     return y
```

Lets look at a sample sigmoid line that might fit with the data:

```python
beta_1 = 0.10
beta_2 = 1990.0

#logistic function
Y_pred = sigmoid(X_data, beta_1 , beta_2)

#plot initial prediction against datapoints
plt.figure(figsize=(8,5))
plt.suptitle(' Plot Sederhana:  Function Sigmoid on data points')
plt.plot(X_data, Y_pred*15000000000000.)
plt.plot(X_data, y_data, 'ro')
plt.show()
```

**Our task here is to find the best parameters for our model. Lets first normalize our x and y:**

```python
# Lets normalize our data
xdata =X_data/max(X_data)
ydata =y_data/max(y_data)
```

```python
max(y_data)
```

#### How we find the best parameters for our fit line?
we can use __curve_fit__ which uses non-linear least squares to fit our sigmoid function, to data. Optimal values for the parameters so that the sum of the squared residuals of sigmoid(xdata, *popt) - ydata is minimized.

popt are our optimized parameters.

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, xdata, ydata)  # pcov is the covariance, popt is the optimized parameters
#print the final parameters
print(" beta_1 = %f, beta_2 = %f" % (popt[0], popt[1]))
```

```python
popt
```

```python
pcov
```

Now we plot our resulting regression model.

```python
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plt.figure(figsize=(8,5))
y = sigmoid(x, popt[0], popt[1])
plt.plot(xdata, ydata, 'ro', label='data')
plt.plot(x,y, linewidth=3.0, label='fit')
plt.legend(loc='best')
plt.ylabel('GDP', color='r', fontsize = 18)
plt.xlabel('Year', color='r', fontsize = 18)
plt.xticks(color ='yellow')
plt.yticks(color = 'y')
plt.show()
```

## Praktek
Can you calculate what is the accuracy of our model?

```python
# split data into train/test
msk = np.random.rand(len(df)) < 0.8
train_x = xdata[msk]
test_x = xdata[~msk]
train_y = ydata[msk]
test_y = ydata[~msk]

# build the model using train set
popt, pcov = curve_fit(sigmoid, train_x, train_y)

# predict using test set
y_hat = sigmoid(test_x, *popt)

# evaluation
mean_abs_error = np.mean(np.absolute(y_hat - test_y))
mean_squ_error = np.mean(np.absolute((y_hat - test_y) **2))

print("Mean absolute error: %.2f" % mean_abs_error)
print("Residual sum of squares (MSE): %.2f" % mean_squ_error)

from sklearn.metrics import r2_score

r_score = r2_score(y_hat, test_y )
print("R2-score: %.2f" % r_score )



```
